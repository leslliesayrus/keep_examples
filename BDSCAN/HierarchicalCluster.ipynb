Hierarchical Cluster

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
import plotly.express as px
import plotly.graph_objects as go
from sklearn.cluster import DBSCAN
```

```python
base = np.array([[20,1200],
[30,2500],
[25,1400],
[22,1400],
[30,3000],
[26,2000],
[40,3500],
[50,4000],
[60,5000],
[33,2600],
[21,1200],
[19,700],
[40,3000],
[45,2500],
[25,2000],
[60,5000],
[23,1900],
[30,2600],
[36,4000]])
```

```python
# normalizacao/padronizacao dos dados 
scaler = StandardScaler()
data = scaler.fit_transform (base)
```

```python
# eps = raio
# samples = numeros de exemplo em uma vizinhanca pra que o ponto faca parte do grpo (default = 5)
dbscan = DBSCAN(eps = 0.5, min_samples = 3)
dbscan.fit(data)
```

```python
# o bdscan classifica outliers como -1
rotulos = dbscan.labels_
```

```python
grafico = px.scatter (x = data[:,0], y = data[:,0], color = rotulos)
grafico.show()
```

```python
from sklearn import datasets
from sklearn.cluster import KMeans 
from sklearn.cluster import AgglomerativeClustering
```

```python
# data grafico
x,y = datasets.make_moons (n_samples = 1500, noise = 0.05)
grafico2 = px.scatter (x=x[:,0], y=x[:,1])
grafico2.show()
```

```python
kmeans = KMeans (n_clusters = 2)
hc = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
dbscan = DBSCAN (eps = 0.1, min_samples = 2)

kmeans.fit(x)
hc.fit(x)
dbscan.fit(x)

r_kmeans, r_hc, r_dbscan = kmeans.labels_,hc.labels_, dbscan.labels_
```

```python
# kmeans - grafico
grafico2 = px.scatter (x=x[:,0], y=x[:,1], color =r_kmeans, title = 'K-MEANS')
grafico2.show()
```

```python
# hierarcjical cluster - grafico
grafico2 = px.scatter (x=x[:,0], y=x[:,1], color =r_hc, title = 'HIERARCHICAL CLUSTER')
grafico2.show()
```

```python
# bdscan - grafico
grafico2 = px.scatter (x=x[:,0], y=x[:,1], color =r_dbscan, title = 'BDSCAN')
grafico2.show()
```
